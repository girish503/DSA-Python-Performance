**Goal:** Optimizing search logic from $O(N^2)$ to $O(N)$ using Hash Maps.

---
### ğŸ§  Core Concept:
Instead of using nested loops, we pre-store data into a **Dictionary** (Hash Map) to achieve constant time $O(1)$ lookup.

```python
# ğŸ Logic: Frequency Mapping using Dictionary
# Based on my handwritten notes dated 24/12/25

def get_frequency_map(nums):
    freq_map = {}
    for i in range(len(nums)):
        if nums[i] in freq_map:
            freq_map[nums[i]] += 1
        else:
            freq_map[nums[i]] = 1
    return freq_map

numbers = [5, 6, 7, 7, 7, 11, 9, 1, 1, 5]
print(f'âœ… Frequency Map Result: {get_frequency_map(numbers)}')
```

### ğŸ”  Character Hashing (a-z)
**Formula:** `index = ord(ch) - 97`

```python
# ğŸ”  Logic: Character Hashing (a-z)
def char_frequency_hashing(text):
    hash_list = [0] * 26
    for ch in text.lower():
        if 'a' <= ch <= 'z':
            index = ord(ch) - 97
            hash_list[index] += 1
    return hash_list

input_str = 'azyxyyzaaaa'
print(f'âœ… Hashed Counts: {char_frequency_hashing(input_str)}')
```

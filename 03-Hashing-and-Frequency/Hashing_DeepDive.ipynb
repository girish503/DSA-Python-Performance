## ðŸš€ Part 2: The Hashing Concept (Pre-storing)

**The Engineering Mindset Shift:**
Instead of searching repeatedly, why don't we scan the data list **once**, count everything, and store it in a "Memory Box" for fast lookups later? This technique is called **Hashing**.

> **Rule of Thumb:** We trade **Space (Memory)** to save **Time**.

### ðŸ”¹ Optimization A: General Frequency Map (Using Dictionaries)

*(Reference: Image 1 & 4 Notes - "Using Dictionary")*

When the numbers can be very large (e.g., billions) or not integers (e.g., strings), we use a **HashMap (Dictionary in Python)**. A dictionary allows us to look up any key in $O(1)$ average time.

**The Logic from Notes:**
1.  Create an empty dictionary `freq_map = {}`.
2.  Iterate through the list.
3.  **Check:** Is the number already a key in the dictionary?
    * **Yes:** Increment its value (+1). (`update the key value`)
    * **No:** Create a new entry and set value to 1. (`write dictionary`)


````python
# Implementation based on Image 1 & 4 Notes
def count_frequencies_dict(nums):
    # Step 1: Initialize empty dictionary
    freq_map = {}
    
    # Step 2: Single pass traversal - O(N)
    # Note loop syntax changed slightly from notes for more Pythonic style, logic remains same.
    for num in nums: 
        # Step 3: Check and Update
        if num in freq_map:
            freq_map[num] += 1
        else:
            freq_map[num] = 1
            
    return freq_map

# Test data from notes
data = [5, 6, 7, 7, 7, 11, 9, 1, 1, 5]
print(f"âœ… Dictionary Result: {count_frequencies_dict(data)}")

# âœ… Complexity Analysis:
# Time: O(N) - We scan the list once.
# Space: O(K) - Where K is the number of unique elements stored in the dict.
````

### ðŸ”¹ Optimization B: Integer Hashing (Using Arrays)

*(Reference: Image 2 & 5 Middle Section - "Optimize hash list")*

**The Interview Trick:** Sometimes, the interviewer gives a crucial constraint: **"All elements in the array are small positive integers (e.g., between 0 and 10)."**

If the range is small, a Dictionary is overkill. We can use a simple **fixed-size Array** as our hash map. This is faster and uses less overhead.

**The Logic from Notes:**
1.  **Identify Max Value:** The notes show elements up to 10. So we need indices 0 to 10.
2.  **Create Array:** Create an array of size 11 (max_value + 1) filled with zeros. `hash_list = [0] * 11`.
3.  **Map Directly:** The number *itself* becomes the index. If we see number `5`, we go to index `5` and increment.
4.  **Formula:** `hash_list[num] += 1`.

````python
# Implementation based on Image 2 & 5 Notes (Integer constraints)
def count_frequencies_array(nums, max_val):
    # Step 1 & 2: Create fixed-size array based on max constraint
    # Notes used size 11 for max element 10.
    hash_list = [0] * (max_val + 1)
    
    # Step 3: Pre-storing (Hashing) - O(N)
    for num in nums:
        # Direct mapping: number is the index
        hash_list[num] += 1
        
    return hash_list

# Test data from notes
n_data = [7, 3, 2, 2, 1, 5, 5, 7, 5, 10]
max_constraint = 10
hashed_array = count_frequencies_array(n_data, max_constraint)

print(f"âœ… Array Hash Result (Index corresponds to number): {hashed_array}")
# Example interpretation: Index 5 has value 3, meaning number 5 appears 3 times.

# âœ… Complexity Analysis:
# Time: O(N) to pre-store. Fetching is O(1).
# Space: O(Max_Value) - Fixed size array regardless of N.
````

### ðŸ”¹ Optimization C: Character Hashing (The ASCII Trick)

*(Reference: Image 2 & 5 Bottom Section - "Character Hashing")*

**The Problem:** Count frequency of lowercase English letters (`'a'` to `'z'`) in a string.

**The Trick:** We can't use characters as array indices directly. We need to map `'a' -> 0`, `'b' -> 1`, ..., `'z' -> 25`. We use ASCII values for this.
* ASCII of `'a'` is 97.
* ASCII of `'b'` is 98.

**The Formula from Notes:**
To get index 0 for 'a': `ord('a') - 97 = 97 - 97 = 0`.
To get index 1 for 'b': `ord('b') - 97 = 98 - 97 = 1`.

**General Formula:** `index = ord(character) - 97` (or `ord(ch) - ord('a')`).

````python
# Implementation based on Image 2 & 5 Notes (Character Hashing)
def char_frequency_hashing(text):
    # Step 1: Create fixed-size array for 26 letters
    # Notes suggest size 27 sometimes for safety, but 26 is exact for a-z.
    hash_list = [0] * 26
    
    # Step 2: Traverse string - O(N)
    for ch in text:
        # Ensure it's a lowercase letter
        if 'a' <= ch <= 'z':
            # Step 3: Apply the ASCII formula from notes
            index = ord(ch) - 97
            hash_list[index] += 1
            
    return hash_list

# Test data from notes
s_data = "azyxyyzaaaa"
char_counts = char_frequency_hashing(s_data)
print(f"âœ… Character Hash Result (a-z): {char_counts}")
# Interpretation: Index 0 holds count of 'a', Index 25 holds count of 'z'.

# âœ… Complexity Analysis:
# Time: O(N) where N is length of string.
# Space: O(1) - Because the array size is always 26, constant space.
````

## ðŸŽ“ Part 3: FAANG/MAANG Interview Questions (The "Why" & "How")

Here are top interview questions based on the concepts from my notes.

### Q1: Why did we move from Brute Force to Hashing? (The Trade-off Question)
**Answer:** In the brute force approach (nested loops), the time complexity is $O(N^2)$, which is too slow for large datasets ($10^8$ elements). By using Hashing, we utilize extra **Space (Memory)** to store counts in a Dictionary or Array. This reduces the time complexity to $O(N)$ for pre-storing and $O(1)$ for fetching. It's a classic **Time-Space Trade-off**.

### Q2: When should I use an Array vs. a Dictionary for hashing?
**Answer:**
* **Use Array:** When the elements are integers and their **range is small and known** (e.g., numbers between 1 and 1000, or characters 'a'-'z'). Array indexing is slightly faster and uses less overhead.
* **Use Dictionary:** When the range is huge (e.g., numbers up to $10^9$), unknown, or when elements are not integers (e.g., strings). Dictionaries handle sparse data efficiently.

### Q3: Explain the logic behind `ord(ch) - 97` in character hashing.
**Answer:** We need to map characters 'a' through 'z' to array indices 0 through 25. Computers store characters as ASCII numbers. 'a' is 97, 'b' is 98, etc. By subtracting the base value of 'a' (97) from any character's ASCII value, we normalize it to start from 0. So, 'a' becomes $97-97=0$, 'b' becomes $98-97=1$. This is a **Direct Addressing Table** technique.

### Q4: What is the worst-case time complexity of Dictionary lookups?
**Answer:** While the *average* case is $O(1)$, the **worst-case** is $O(N)$. This happens due to "Collisions" (when many different keys hash to the same underlying bucket). However, modern languages (like Python) handle this very efficiently, so we usually assume $O(1)$ in interviews unless specific constraints force a collision scenario.

### Q5: Can we optimize space in Character Hashing if the string only has 'a', 'b', and 'c'?
**Answer:** Yes. The current logic uses $O(26)$ constant space. If we know for sure only 'a', 'b', 'c' exist, we could use an array of size 3. However, $O(26)$ is already considered $O(1)$ constant space in asymptotic analysis, so further optimization is usually not required unless memory is extremely tight.

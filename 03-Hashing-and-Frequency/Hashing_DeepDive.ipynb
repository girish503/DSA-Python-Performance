### üß† The Problem Statement
Suppose we have a list of numbers or characters. We need to find how many times each element appears. 

**Example Input:** `[5, 6, 7, 7, 11, 9, 1, 1, 5]`
**Expected Output:** `{5: 2, 6: 1, 7: 2, 11: 1, 9: 1, 1: 2}`

## üõ†Ô∏è Approach 1: Brute Force (The Native Way)
Krish Naik style logic: Modata manam technology lekunda oka saamaanya manishi laa alochisthe, nested loops vadatham.

### üí° Telugu Explanation:
1. Oka element ni pattuko, motham array lo vethiki counts count chey.
2. Malli rendava element ni pattuko, malli motham array vethuku.
3. Idhi nested loops (for loop lopala for loop) thone saadhyam.

### ‚öñÔ∏è Complexity Analysis:
* **Time Complexity (TC):** $O(N^2)$ (Very Slow - Matrix calculations level logic)
* **Space Complexity (SC):** $O(1)$ (No extra memory used)

## üöÄ Approach 2: Optimal Hashing (The Professional Way)
Ikkada manam **Hashing** ane concept vadatham. Instead of searching again and again, manam **Memory (Space)** ni use chesi **Time** ni save chestham.

### üí° The 'Aha!' Moment (Tricks):
* **Trick 1:** Use a Dictionary as a 'Memory Box'.
* **Trick 2:** Lookup in a Dictionary is $O(1)$ constant time. 
* **Trick 3:** Use `hash_map.get(item, 0)` to handle cases where keys don't exist yet.

```python
# üêç Implementation: Frequency Mapping with Dictionary
def get_optimal_frequency(nums):
    # TC: O(N) | SC: O(N)
    freq_map = {} 
    
    for num in nums:
        # The Hashing Trick: Check and update in one go
        if num in freq_map:
            freq_map[num] += 1
        else:
            freq_map[num] = 1
            
    return freq_map

data = [5, 6, 7, 7, 11, 9, 1, 1, 5]
print(f'‚úÖ Optimal Frequency: {get_optimal_frequency(data)}')
```

## üî† Character Hashing (Interview Special)
